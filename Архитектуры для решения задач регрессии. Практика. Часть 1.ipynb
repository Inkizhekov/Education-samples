```python
import torch
from torch.autograd import Variable

import matplotlib.pyplot as plt
from IPython.display import IFrame
%matplotlib inline

import numpy as np
import imageio

torch.manual_seed(42);
```

```python
tensor = torch.rand(100) #torch.rand(100) делаем переменную куда жраним созданный массив
```

```python
tensor.sort()[0] #torch.sort(torch.rand(100))[0] работаем с уже созданнвм массивом, а не геренируем новый
```

```python
x = torch.unsqueeze(torch.sort(tensor)[0], dim=1)
x
```

```python
y_true = 2*x
y = y_true + (torch.rand(x.size()) - 0.5)
y
```

```python
x, y = Variable(x), Variable(y)

plt.figure(figsize=(10,4))
plt.scatter(x.data.numpy(), y.data.numpy(), color = "black", label = "Измерения", s = 15)
plt.plot(x.data.numpy(), y_true, color = "red", label = "y")
plt.title('Пример зависимости с шумом', fontdict = {'size': 15})
plt.xlabel('x', fontdict = {'size': 15})
plt.ylabel('y', fontdict = {'size': 15})
plt.xlim(-0.05, 1.01)
plt.ylim(-0.55, 2.55)
plt.legend()
plt.show()
```

```python
class LinearNet(torch.nn.Module):
  def __init__(self, n_feature, n_output, use_bias=False):
    super(LinearNet, self).__init__()
    self.output = torch.nn.Linear(n_feature, n_output, bias=use_bias)

  def forward(self, x):
    x = self.output(x)
    return x
```

```python
net = LinearNet(n_feature=1, n_output=1)
print(net)
```

```python
loss_func = torch.nn.MSELoss()
```

```python
prediction = net(x)
loss = loss_func(prediction, y)
```

```python
prediction
```

```python
loss
```

```python
torch.mean((prediction - y)**2)
```

```python
optimizer = torch.optim.SGD(net.parameters(), lr=0.2)

my_images= []
fig, ax = plt.subplots(figsize=(12,7))

for t in range (100):
  prediction = net(x)
  loss = loss_func(y, prediction)

  optimizer.zero_grad()
  loss.backward()
  optimizer.step()

  plt.cla()
  ax.scatter(x.data.numpy(), y.data.numpy(), color = "black", label = "Измерение", s=15)
  ax.plot(x.data.numpy(), y_true, color = "red", label = "y")
  ax.plot(x.data.numpy(), prediction.data.numpy(), color = "blue", label = "Предсказание")

  ax.set_title('Построение регресии', fontdict={'size':15})
  ax.set_xlabel('x', fontdict={'size':15})
  ax.set_ylabel('y', fontdict={'size':15})
  ax.set_xlim(-0.05, 1.05)
  ax.set_ylim(-0.55, 2.55)
  ax.legend()

  fig.canvas.draw()
  image = np.frombuffer(fig.canvas.tostring_rgb(), dtype='uint8')
  image = image.reshape(fig.canvas.get_width_height()[::-1]+ (3,))

  my_images.append(image)

for _ in range(10):
  my_images.append(image)

imageio.mimsave('.gif/simple.gif', my_images, fps=10)
IFrame(src='.gif/simple.gif', width=100, height=550)
```
