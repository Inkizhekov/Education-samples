# Создаем датафрейм

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
```

```python
city_list = ['Москва', 'Санкт-Петербург', 'Новосибирск', 'Екатеринбург',
             'Казань', 'Нижний Новгород']
popul_list = [12600000, 5400000, 1625000, 1500000, 1300000, 1250000]

index_list = [1, 2, 3, 4, 5, 6]

```

```python
df = pd.DataFrame(columns = ['city', 'population'], index = index_list)

df.head()
```

```python
df['city'] = city_list
df['population'] = popul_list

df.head()
```

### LabelEncoder

```python
labelencoder = LabelEncoder()

df['city_label'] = labelencoder.fit_transform(df['city'])
df.head()
```

```python
labelencoder.inverse_transform([2,4,1])
```

### OneHotEncoding

```python
ohe = OneHotEncoder()
```

```python
df_ohe = pd.DataFrame(ohe.fit_transform(df[['city_label']]).toarray())

df = df.join(df_ohe)
```

```python
df
```

### Pandas get_dummies

```python
dm_df = pd.get_dummies(df, columns = ['city_label'], prefix = ['get_D'])
dm_df
```
