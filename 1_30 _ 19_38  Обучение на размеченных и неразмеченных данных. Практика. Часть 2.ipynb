```python
import numpy as np # для работы с массивами
import pandas as pd #для создания DataFrame

from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

import matplotlib.pyplot as plt
```

```python
np.random.seed(40)
```

```python
X, y = make_blobs(
    n_samples=200, n_features=2, 
    centers=4, cluster_std=0.7, 
    shuffle=True
)

plt.title("Сгенерированные точки")
plt.scatter(X[:,0], X[:,1], c = y)
plt.show
```

```python
plt.title("Сгенерированные точки")
plt.scatter(X[:,0], X[:,1])
plt.show
```

```python
kmeans = KMeans(n_clusters=4, init='random', n_init=10, max_iter=100, tol=1e-04)
kmeans_prediction = kmeans.fit_predict(X)
```

```python
kmeans_prediction
```

```python
y
```

```python
plt.title("Сгенерированные точки")
plt.scatter(X[:,0], X[:,1], c = y)
plt.show
```

```python
plt.title('Результат кластеризации')
plt.scatter(X[:,0], X[:,1], c = kmeans_prediction)
plt.scatter(*kmeans.cluster_centers_.T, marker="*",s=200, color='r')
```

```python
distortion = []
for i in range(1, 7):
  kmeans = KMeans(n_clusters=i, init='random', n_init=10, max_iter=100, tol=1e-04)
  kmeans_prediction = kmeans.fit_predict(X)
  plt.title('Результат кластеризации')
  plt.scatter(X[:,0], X[:,1], c = kmeans_prediction)
  plt.scatter(*kmeans.cluster_centers_.T, marker="*",s=200, color='r')
  plt.show()
  distortion.append(kmeans.inertia_)
```

Метод локтя

```python
plt.plot(range(1,7), distortion, marker='o')
plt.xlabel('Количество кластеров')
plt.ylabel('Разброс')
plt.show
```
