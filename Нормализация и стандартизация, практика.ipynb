```python
#импорт
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn. preprocessing import MinMaxScaler
```

```python
#загружаем данные
url = "https://raw.githubusercontent.com/vega/vega/master/docs/data/seattle-weather.csv"
```

```python
#dataset
dataset = pd.read_csv(url)
dataset.head()
```

```python
dataset.describe()
```

```python
# minmax scaler
minmax_scaler = MinMaxScaler()
#Обучение MinMaxScaler
dataset_norm = minmax_scaler.fit_transform(np.array(dataset[['precipitation', 'temp_max', 'temp_min', 'wind']]))
```

```python
dataset_norm[1]
```

```python
#возвращаем название столбцов (преобразование обратно в датафрейм)
dataset_norm_df = pd.DataFrame(dataset_norm, columns = ['precipitation', 'temp_max', 'temp_min', 'wind'])
dataset_norm_df.head()
```

```python
dataset_norm_df.describe()
```

```python
# standartscaler

from sklearn.preprocessing import StandardScaler
```

```python
#scaler
std_scaler = StandardScaler()
dataset_std = std_scaler.fit_transform(np.array(dataset[['precipitation', 'temp_max', 'temp_min', 'wind']]))

dataset_std_df = pd.DataFrame(dataset_std, columns = ['precipitation', 'temp_max', 'temp_min', 'wind'])
dataset_std_df.head()
```

```python
dataset_std_df.describe()
```
